```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
```

## Data Import

```python
df = pd.read_csv("../data/sentences.csv")
```

```python
r,c = df.shape

print(f"Le nombre de ligne : {r}")
print(f"Le nombre de colonne : {c}")

df.head()
df = df.dropna()
print(f"Nombre de valeurs manquantes\n{df.isnull().sum()}")

eng = df["eng"].str.len() 
darija_ar = df["darija_ar"].str.len()
darija = df["darija"].str.len()

plt.hist(eng,label="eng")
plt.hist(darija,label="darija")
plt.hist(darija_ar,label="darija_ar")
plt.legend()

df = df.drop(columns = ["darija"])
r,c = df.shape

print("==== Apres nettoyage ====")

print(f"Le nombre de ligne : {r}")
print(f"Le nombre de colonne : {c}")
```

## Train/Validation/test split

```python
X_train, X_test, y_train, y_test = train_test_split(df["eng"],df["darija_ar"], test_size = 0.1)
X_train, X_val, y_train, y_val = train_test_split(X_train,y_train, test_size = 0.1)

```

```python
print(f"Shape Xtrain {X_train.shape}")
print(f"Shape Xval {X_val.shape}")
print(f"Shape yval {y_val.shape}")
print(f"Shape y_train {y_train.shape}")
print(f"Shape Xtest {X_test.shape}")
print(f"Shape ytest {y_test.shape}")

train = [X_train,y_train]
val = [X_val,y_val]
test = [X_test,y_test]

df_train = pd.concat(train,axis = 1)
df_val = pd.concat(val,axis = 1)
df_test = pd.concat(test,axis = 1)


df_train.to_csv("../data/Train.csv",index = False)
df_val.to_csv("../data/Val.csv",index = False)
df_test.to_csv("../data/Test.csv",index = False)
```
